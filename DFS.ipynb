# DFS(깊이 우선 탐색)

- 비선형구조인 그래프 구조는, 그래프로 표현된 모든 자료를 빠짐 없이 검색하는 것이 point!
    - DFS 깊이 우선 탐색
    - BFS 너비 우선 탐색
    
- 한쪽 방향으로 계속 탐색하다가, 더 이상 진행 할 수 없으면, 다시 되돌아 오는 방법으로 탐색했다는 것을 확인

- 다시 되돌아 오기 위해 사용한 DS 로 stack(lst) 사용

---

## DFS 처리 방식

- 시작 정점의 한 방향으로 갈 수 있는 경로가 있는 곳 까지 깊이 탐색

- 더 이상 갈 곳이 없게 되면, 가장 마지막에 만났던 갈림길 간선이 있는 정점으로 되돌아옴

- 다른 방향의 정점으로 탐색을 게속 반복하여, 결국 모든 정점을 방문하게 되며 순회

- 가장 마지막에 만났던 갈림길의 정점으로 되돌아가서, 다시 DFS(깊이 우선 탐색)을 반복해야 하므로, LIFO(후입 선출) 구조의 stack(lst) 데이터 구조 사용

---

## DFS 알고리즘

1. 시작 정점 S를 결정하여 방문,

2-1) 정점 S에 인접한 정점 중에서, 방문하지 않은 정점 w가 있으면, 
    - 정점 S를 stack에 append() 하고, 방문하지 않은 정점 w를 방문
    - w를 정점 S(V) 로 해서 다시 반복
    
2-2) 정점 S에 인접한 정점 중에서, 방문하지 않은 정점 w가 없으면,
    - 탐색의 방향을 바꾸기 위해서, stack(lst)을 pop하여, 이전에 append() 해놓은 가장 마지막 방문 정점을 V로 하여, 다시 반복

```python
def dfs(c):
    
    # 2-1) 현재 노드 방문하면서 방문 표시
    v[c] = 1 
    
    # 2-3) 방문 처리 한 c를 정답 처리 
    ans.append(c) 
    
    # 2-2) 정점 c 기준, 인접한 정점들 중에서, 미방문한 정점이 있다면, 방문, DFS 탐색 
    for n in adj[c]:
        
        if v[n] == 0:
            
            dfs(n) 
            
```
